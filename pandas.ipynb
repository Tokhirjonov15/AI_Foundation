```python
data_with_missing = {
    "cars": ["BMW", "Volvo", None, "Tesla", "Kia", None],
    "counts": [3, 7, None, 8, None, 4]
}

df_missing = pd.DataFrame(data_with_missing)
print(df_missing)
```

```python
df_missing.isnull().sum()
```

```python
df_missing.dropna(how="all")   # (axis=1) agar columnda Falsy value bo'lsa columnni butunlat o'chiradi. # (subset=['counts']) bu faqat counts columnni tekshirib beradi
```

```python
df_missing.fillna({"cars": "unknowm", "counts": 0}) # Falsy valuelarga qiymat qo'shish
```
