```python
import numpy as np

arr = np.array([1, 3, 5, 7])
print("Numpy Array:", arr)

arr.shape
```

```python
arr2 = np.array([[1, 3, 5, 7], [4, 6, 8, 9]])
print("Numpy Array2:", arr2)

arr2.shape
```

```python
# dtype()
a = np.array([1, 2, 3])
print(a.dtype)
```

```python
# reshape()
a = np.array([1, 2, 3, 4, 5, 6])
b = a.reshape(2, 3)
print("shaped_array:", b)
```

```python
# astype()
a = np.array([1, 2, 3])
b = a.astype(float)
print("float:", b)
```

```python
# np.zeros(), np.ones() => dummy data
z = np.zeros((2, 3))
o = np.ones((2, 4))

print("Zeros:", z)
print("Ones:", o)
```

```python
# ndim
a = np.array([[1, 2], [3, 4]])
print("ndim:", a.ndim)
```

```python
# arange() making array
a = np.arange(1, 7)
print("Arange:", a)
```

```python
# sum() yig'indi hisoblaydi, mean() O'rtachani hisoblaydi, max() eng kattasi, min() eng kichigi
a = np.array([1, 2, 3, 4])

print("Sum:", a.sum())
print("Mean:", a.mean())
print("Max:", a.max())
print("Min:", a.min())
```

```python
# flatten() 2D yoki 3Dni 1Dga aylantiradi

a = np.array([[1, 2], [3, 4]])
print("Flatten:", a.flatten())
```

```python
# T (Transpose)
a = np.array([[1, 2], [3, 4]])
print("Transpose:", a.T)

```

```python
# dot() @
a = np.array([[1, 2], [3, 4]])
b = np.array([[2, 0], [1, 2]])
print("Dot/Multiplictaion:", a.dot(b))
```

```python
# random.rand()  0dan 1gacha sonlardan array yaratadi
a = np.random.rand(2, 3)
print("Random Rand:", a)
```
